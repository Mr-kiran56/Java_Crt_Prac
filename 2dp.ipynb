```python
from typing import List
w=[2,1,3,4]
val=[5,2,8,4]
W=5

def knapSack(w:List,val:List,W:int)->int:
    n=len(w)
    dp=[[0]*(W+1) for _ in range(n+1)]
    
    for i in range(n+1):
        for j in range(W+1):
            if (w[i-1]<=j):
                dp[i][j]=max(dp[i-1][j],val[i-1]+dp[i-1][j-w[i-1]])
            else:
                dp[i][j]=dp[i-1][j]
    return dp[n][W]
print(knapSack(w,val,W))
                
    
```

```python
from typing import List
arr=[1,1,1]
k=3

def countSums(arr:List,k:int)->int:
    n=len(arr)
    dp=[[0]*(k+1) for _ in range(n+1)]
    for i in range(n+1):
        dp[i][0]=1
    for i in range(1,n+1):
        for j in range(1,k+1):
            if (arr[i-1]<=j):
                dp[i][j]=dp[i-1][j]+dp[i-1][j-arr[i-1]]
            else:
                dp[i][j]=dp[i-1][j]
    print(dp)
    return dp[n][k]

print(countSums(arr,k))
```

```python
text1 = "abac"
text2 = "caba"


def lcs(text1:str,text2:str)->int:
    n=len(text1)
    m=len(text2)
    dp=[[0]*(m+1) for _ in range(n+1)]
    for i in range(1,n+1):
        for j in range(1,m+1):
            if text1[i-1]==text2[j-1]:
                dp[i][j]=1+dp[i-1][j-1]
            else:
                dp[i][j]=max(dp[i-1][j],dp[i][j-1])
                
    return dp[n][m]
print(lcs(text1,text2))
```

```python
nums = [10,9,2,5,3,7,101,18]
inums=[]
inums.append(nums[0])
def lis(nums:List,inums:List)->List:
    n=len(nums)
    for i in range(1,n):
        if inums[-1]>nums[i]:
            inums[-1]=nums[i]
        if inums[-1]<nums[i] and inums[-1]!=nums[i]:
            inums.append(nums[i])
    return len(inums)
print(lis(nums,inums))
```

```python
nums = [7,7,7,7,7,7,7]

def lis(nums:List)->List:
    n=len(nums)
    dp=[1]*n
    for i in range(n):
        for j in range(i):
            if(nums[j]<nums[i]):
                dp[i]=max(dp[i],dp[j]+1)
    return dp[-1]
print(lis(nums))
```

#Edit Distance

```python
s1='abedf'
s2='adebf'

def editDistance(s1:str,s2:str)->int:
    n1=len(s1)
    n2=len(s2)
    dp=[[0]*(n1+1) for _ in range(n2+1)]
    
    for i in range(n1+1):
        dp[i][0]=i
    for j in range(n2+1):
        dp[0][j]=j
    
    
    for i in range(1,n1+1):
        for j in range(1,n2+1):
            if s1[i-1]==s2[j-1]:
                dp[i][j]=dp[i-1][j-1]
            else:
                dp[i][j]=min(dp[i-1][j-1],dp[i-1][j],dp[i][j-1])+1
    print(dp)
    return dp[n1][n2]


editDistance(s1,s2)
```

#Wild Card Pattern 

```python
expreesion='abcd'
pattern='a?*d'

def wildcardPattern(expreesion:str,pattern:str)->bool:
    n=len(expreesion)
    m=len(pattern)
    dp=[[False]*(m+1) for _ in range(n+1)]
    dp[0][0]=True
    
    for i in range(n+1):
        for j in range(1,m+1):
            
            if i==0:
                dp[i][j]=(j>0 and dp[i][j-1] or j==0) and (pattern[j-1]=='*')
            
            elif pattern[j-1]=='?' or expreesion[i-1]==pattern[j-1]:
                dp[i][j]=dp[i-1][j-1]
                
            elif pattern[j-1]=='*':
                dp[i][j]=dp[i-1][j] or dp[i][j-1]
    return dp[n][m]
print(wildcardPattern(expreesion,pattern))
```

#Regular Expression

```python
expreesion='abcde'
pattern='a*b.de'

def regularExpreesion(expreesion:str,pattern:str)->bool:
    n=len(expreesion)
    m=len(pattern)
    dp=[[False]*(m+1) for _ in range(n+1)]
    dp[0][0]=True
    for i in range(n+1):
        for j in range(1,m+1):
            
            if i==0:
                dp[i][j]=(j>0 and dp[i][j-1] or j==0) and pattern[j-1]=='*'
            elif pattern[j-1]=='.' or pattern[j-1]==expreesion[i-1]:
                dp[i][j]=dp[i-1][j-1]
            elif pattern[j-1]=='*':
                dp[i][j]=dp[i-1][j] or dp[i][j-1]
    return dp[n][m]

print(regularExpreesion(expreesion,pattern))
```
